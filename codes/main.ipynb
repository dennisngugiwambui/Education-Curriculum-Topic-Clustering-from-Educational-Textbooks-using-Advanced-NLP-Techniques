- Hierarchical Agglomerative Clustering

```python
# RESULTS VISUALIZATION AND ANALYSIS - UPDATED CELL 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Configure matplotlib for Jupyter
%matplotlib inline
plt.rcParams['figure.figsize'] = (14, 10)
plt.rcParams['font.size'] = 10

try:
    # Load latest results with better error handling
    result_files = list(Path('.').glob('masters_research_*.json'))
    
    # Filter out dataset and stats files, prioritize comprehensive reports
    filtered_files = []
    comprehensive_files = []
    
    for f in result_files:
        if 'dataset' not in f.name and 'stats' not in f.name:
            if 'comprehensive_report' in f.name:
                comprehensive_files.append(f)
            else:
                filtered_files.append(f)
    
    # Use comprehensive report if available, otherwise use other files
    target_files = comprehensive_files if comprehensive_files else filtered_files
    
    if target_files:
        latest_results = max(target_files, key=lambda x: x.stat().st_mtime)
        
        print(f"üìä RESEARCH RESULTS SUMMARY")
        print(f"Report: {latest_results.name}")
        
        # Load with UTF-8 encoding
        with open(latest_results, 'r', encoding='utf-8') as f:
            research_data = json.load(f)
        
        # Extract metadata
        metadata = research_data.get('research_metadata', {})
        print(f"Execution Time: {metadata.get('execution_duration', 'N/A')}")
        print(f"Phases Completed: {metadata.get('total_phases_completed', 0)}/4")
        print(f"Errors Encountered: {metadata.get('errors_encountered', 0)}")
        
        # Extract methodology data
        methodology = research_data.get('methodology', {})
        data_collection = methodology.get('data_collection', {})
        eval_metrics = methodology.get('evaluation_metrics', {})
        nlp_techniques = methodology.get('nlp_techniques', {})
        
        print(f"\nüìà KEY PERFORMANCE METRICS:")
        print(f"Documents Processed: {data_collection.get('total_documents', 0):,}")
        print(f"Topic Coherence Score: {eval_metrics.get('topic_coherence', 0):.4f}")
        print(f"Topic Modeling: {nlp_techniques.get('topic_modeling', 'N/A')}")
        print(f"Clustering Method: {nlp_techniques.get('clustering', 'N/A')}")
        
        # Extract results data
        results = research_data.get('results', {})
        if results:
            data_results = results.get('data_collection_results', {})
            print(f"Items Scraped: {data_results.get('total_items_scraped', 0):,}")
            print(f"Quality Items Retained: {data_results.get('quality_items_retained', 0):,}")
            print(f"Collection Status: {data_results.get('status', 'N/A')}")
        
        # Research objectives status
        objectives = research_data.get('research_objectives', {})
        if objectives:
            achieved = objectives.get('achieved', False)
            status_icon = "‚úÖ" if achieved else "‚ùå"
            print(f"\nüéØ RESEARCH OBJECTIVES: {status_icon} {'ACHIEVED' if achieved else 'PENDING'}")
            print(f"Primary: {objectives.get('primary', 'N/A')}")
        
        # Create visualization
        print(f"\nüìä Creating Research Visualization...")
        
        # Set up the plots
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Masters Research: Curriculum Topic Clustering Results', 
                    fontsize=16, fontweight='bold', y=0.98)
        
        # Plot 1: Data Collection Results
        scraped = data_results.get('total_items_scraped', 0) if results else 0
        retained = data_results.get('quality_items_retained', 0) if results else 0
        
        categories = ['Items\nScraped', 'Quality\nRetained']
        values = [scraped, retained]
        colors = ['#3498db', '#2ecc71']
        
        bars1 = ax1.bar(categories, values, color=colors, alpha=0.8, edgecolor='white', linewidth=2)
        ax1.set_title('Data Collection Results', fontweight='bold', pad=20)
        ax1.set_ylabel('Number of Documents')
        ax1.grid(axis='y', alpha=0.3)
        
        # Add value labels
        for bar, value in zip(bars1, values):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + max(values)*0.01,
                    f'{value}', ha='center', va='bottom', fontweight='bold', fontsize=12)
        
        # Plot 2: Topic Coherence Score
        coherence = eval_metrics.get('topic_coherence', 0)
        coherence_data = [coherence, 1.0 - coherence]
        labels = [f'Coherence\n{coherence:.3f}', f'Remaining\n{1-coherence:.3f}']
        colors2 = ['#e74c3c', '#ecf0f1']
        
        wedges, texts, autotexts = ax2.pie(coherence_data, labels=labels, colors=colors2, 
                                          autopct='%1.1f%%', startangle=90, 
                                          textprops={'fontweight': 'bold'})
        ax2.set_title(f'Topic Coherence Analysis', fontweight='bold', pad=20)
        
        # Plot 3: NLP Configuration
        # Extract numbers from strings
        topics_str = nlp_techniques.get('topic_modeling', 'LDA with 30 topics')
        clusters_str = nlp_techniques.get('clustering', 'Hierarchical clustering with 25 clusters')
        
        topics = 30  # default
        clusters = 25  # default
        
        # Try to extract actual numbers
        try:
            import re
            topic_match = re.search(r'(\d+)\s*topics?', topics_str, re.IGNORECASE)
            if topic_match:
                topics = int(topic_match.group(1))
            
            cluster_match = re.search(r'(\d+)\s*clusters?', clusters_str, re.IGNORECASE)
            if cluster_match:
                clusters = int(cluster_match.group(1))
        except:
            pass
        
        config_labels = ['LDA\nTopics', 'Hierarchical\nClusters']
        config_values = [topics, clusters]
        colors3 = ['#9b59b6', '#f39c12']
        
        bars3 = ax3.bar(config_labels, config_values, color=colors3, alpha=0.8, 
                       edgecolor='white', linewidth=2)
        ax3.set_title('NLP Analysis Configuration', fontweight='bold', pad=20)
        ax3.set_ylabel('Count')
        ax3.grid(axis='y', alpha=0.3)
        
        # Add value labels
        for bar, value in zip(bars3, config_values):
            height = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2., height + max(config_values)*0.01,
                    f'{value}', ha='center', va='bottom', fontweight='bold', fontsize=12)
        
        # Plot 4: Execution Summary
        phases = metadata.get('total_phases_completed', 0)
        errors = metadata.get('errors_encountered', 0)
        max_phases = 4
        
        summary_labels = [f'Phases\nCompleted', f'Errors\nEncountered']
        summary_values = [phases, errors]
        colors4 = ['#2ecc71', '#e74c3c']
        
        bars4 = ax4.bar(summary_labels, summary_values, color=colors4, alpha=0.8,
                       edgecolor='white', linewidth=2)
        ax4.set_title(f'Pipeline Execution Summary\nDuration: {metadata.get("execution_duration", "N/A")}', 
                     fontweight='bold', pad=20)
        ax4.set_ylabel('Count')
        ax4.set_ylim(0, max(max_phases + 1, max(summary_values) + 1))
        ax4.grid(axis='y', alpha=0.3)
        
        # Add value labels
        for bar, value in zip(bars4, summary_values):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{value}', ha='center', va='bottom', fontweight='bold', fontsize=12)
        
        # Add phase completion indicator
        ax4.text(0, phases + 0.3, f'{phases}/{max_phases}', ha='center', va='bottom', 
                fontweight='bold', fontsize=10, color='darkgreen')
        
        plt.tight_layout()
        plt.subplots_adjust(top=0.93)
        plt.show()
        
        # Final summary
        print(f"\nüéâ PIPELINE STATUS: {'‚úÖ SUCCESS' if phases == 4 and errors == 0 else '‚ö†Ô∏è PARTIAL'}")
        print(f"üìÅ Results saved in: {latest_results.name}")
        print(f"üöÄ Ready for thesis writing and academic presentation!")
        
    else:
        print("‚ö†Ô∏è No valid results files found.")
        print("Available JSON files:")
        all_json = list(Path('.').glob('*.json'))
        for f in all_json[:5]:
            print(f"   - {f.name}")
        print("\nRun the main pipeline first: python main.py")

except FileNotFoundError as e:
    print(f"‚ùå File not found: {e}")
    print("Make sure you're in the correct directory with the results files.")
    
except json.JSONDecodeError as e:
    print(f"‚ùå JSON parsing error: {e}")
    print("The results file may be corrupted. Try running the pipeline again.")
    
except Exception as e:
    print(f"‚ùå Unexpected error: {e}")
    print(f"Error type: {type(e).__name__}")
    print("Check that all required packages are installed and try again.")
```
